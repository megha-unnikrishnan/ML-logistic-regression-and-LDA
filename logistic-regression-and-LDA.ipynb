```python
import pandas as pd
```

```python
import numpy as np
```

```python
import seaborn as sn
```

```python
df=pd.read_csv('Movie_collection.csv')
```

```python
df.head()
```

```python
#Training logistic model
```

```python
x=df.loc[:,df.columns!='Start_Tech_Oscar'] #Independent variable
```

```python
y=df['Start_Tech_Oscar'] #dependent variable
```

```python
x.head()
```

```python
y.head()
```

```python
df=pd.get_dummies(df) #converts series of strings to dummy codes
df.info() #provides the information about the dataset
df.Time_taken=df.Time_taken.fillna(df.Time_taken.mean()) #fill the missing values by replacing mean
df.fillna(0) #fill the NaN values with 0
```

```python
#Use Logistic regression model to predict
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
cls_rgs=LogisticRegression()  #Create classification object
```

```python
#Fit the trained data
```

```python
cls_rgs.fit(x,y)
```

```python
# coefficients
```

```python
cls_rgs.coef_ #Î²0
```

```python
cls_rgs.intercept_ #Î²1
```

```python
#use add_constant
```

```python
import statsmodels.api as sn
```

```python
x_const=sn.add_constant(x) #By default, statsmodels fits a line passing through the origin,
                           #i.e. it doesn't fit an intercept. Hence, you need to use the command 'add_constant'
                           #so that it also fits an intercept.
```

```python
x_const
```

```python
# building the model and fitting the data
```

```python
import statsmodels.discrete.discrete_model as sm
```

```python
logit=sm.Logit(y,x_const).fit()
```

```python
#In the output, â€˜Iterationsâ€˜ refer to the number of times the model iterates over the data, trying to optimize the model. By default, the maximum number of iterations performed is 58, after which the optimization fails
```

```python
logit.summary() #The summary table below gives us a descriptive summary about the regression results.
```

```python
# performing predictions on the test dataset
# The predict() function is useful for performing predictions. The predictions obtained are fractional values
# (between 0 and 1) which denote the probability of getting admitted. These values are hence rounded, to obtain 
# the discrete values of 1 or 0. 
```

```python
x_prob=cls_rgs.predict_proba(x) #predict the probabality of x
```

```python
x_prob
```

```python
y_pred=cls_rgs.predict(x) #predicted values
```

```python
y_pred
```

```python
# Using Confusion Matrix to find the Accuracy of the model
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y,y_pred)
```
